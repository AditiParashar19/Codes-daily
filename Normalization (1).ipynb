# Normalization

```python
import pandas as pd 
data = { 
'Student': ['Amit', 'Neha', 'Rahul', 'Priya', 'Karan'], 
'Marks': [45, 88, 76, 95, 60] 
} 
df = pd.DataFrame(data) 
print("Original Data:") 
print(df)
```

# Min-Max Normalization

```python
df['Marks_MinMax'] = (df['Marks'] - df['Marks'].min())/ (df['Marks'].max() - df['Marks'].min())
print(df)
```

# Z-Score

```python
marks = [45,88,76,95,60]
df = pd.DataFrame({
    'Marks':marks
})
mean = df['Marks'].mean()
std = df['Marks'].std()
#z-score
print("Mean",mean)
print("Standard deviation",std)
df['Z_score'] = (df['Marks']-mean)/std
print(df)
```

# Decimal Scaling Normalization

```python
df['Marks_Decimal'] = df['Marks']/100
print(df)
```

# Max Normalization

```python
df['Marks_MaxNorm'] = df['Marks']/df['Marks'].max()
print(df)
```
