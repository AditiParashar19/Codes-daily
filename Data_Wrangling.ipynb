# Data Wrangling

# Create Raw dataset

```python
import pandas as pd 
import numpy as np 
data = { 
    'Employee': ['Aman', 'Riya', 'Kunal', 'Riya', 'Vikas', 'Meena'], 
    'Salary': [50000, np.nan, 45000, np.nan, 999999, 40000], 
    'Experience': [5, 3, None, 3, 8, 2], 
    'Age': ['25', '28', '30', '28', '35', 'thirty'] 
} 
df = pd.DataFrame(data) 
print("RAW DATA:") 
print(df) 
```

```python
df.isnull().sum()
```

# Handle Missing Values

```python
df['Salary'].fillna(df['Salary'].mean(),inplace=True)
df['Experience'].fillna(df['Experience'].mean(),inplace=True)
print(df)
```

```python
df.info()
```

```python
df.describe()
```

# Remove Duplicate Records

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Age'] = pd.to_numeric(df['Age'],errors='coerce')
print(df)
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
print(df)
```

# Handle Outliers

```python
df = df[df['Salary']<200000]
print(df)
```

# Data Transformation

```python
df['Level'] = df['Experience'].apply(lambda x:'Senior' if x>=5 else 'Junior')
print(df)
```

# Data Normalization - Min-Max Normalization

```python
df['Salary_Normalized']=(df['Salary']-df['Salary'].min())/(df['Salary'].max()-df['Salary'].min())
print(df)
```
