# Smoothing Techniques

- 1) moving average 
- 2) exponential smoothing


# Moving Average

```python
import pandas as pd 
data = {'Temperature': [30, 31, 29, 45, 30, 32, 31]} 
df = pd.DataFrame(data) 
df['Moving_Avg'] = df['Temperature'].rolling(window=3).mean() 
print(df) 
```

# Binning

```python
# sort the data
# divide into bins
# replace values with bin mean/median
```

```python
import numpy as np
sorted_data = np.sort(df['Temperature']) 
bins = np.array_split(sorted_data, 3) 
bin_means = [np.mean(bin) for bin in bins] 
print("Bins:", bins) 
print("Bin Means:", bin_means)
```

# Regression

```python
import numpy as np 
from sklearn.linear_model import LinearRegression 
X = np.arange(len(df)).reshape(-1, 1) 
y = df['Temperature'] 
model = LinearRegression() 
model.fit(X, y) 
df['Trend_Line'] = model.predict(X) 
print(df)
```
