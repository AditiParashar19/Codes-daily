```python
df['Salary'].fillna(df['Salary'].median(), inplace=True)
df['Age'].fillna(df['Age'].mean(), inplace=True)
df.head()
```

```python
# Convert Joining_Date to datetime
df['Joining_Date'] = pd.to_datetime(df['Joining_Date'])
df.head()
```

# Handling Outliers

```python
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Salary'] < lower) | (df['Salary'] > upper)]
```

```python
# removing outliers
df = df[(df['Salary'] >= lower) & (df['Salary'] <= upper)]
df.head()
```

# Detect Outliers Using Z-score

```python
# Calculate Z-score
df['Salary_zscore'] = (
    df['Salary'] - df['Salary'].mean()
) / df['Salary'].std()

df.head()
```

```python
outliers = df[abs(df['Salary_zscore']) > 3]
```

```python
# remove outliers
df = df[abs(df['Salary_zscore']) <= 3]
df.head()
```

# Log Tranformation

```python
df['Salary_log'] = np.log1p(df['Salary'])
df.head()
```

# Square Root Transformation

```python
df['Salary_sqrt'] = np.sqrt(df['Salary'])
df.head()
```

# Min-Max Normalization

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df['Salary_MinMax'] = scaler.fit_transform(df[['Salary']])

df.head()
```

# Decimal Normalization

```python
j = len(str(int(df['Salary'].abs().max())))

# decimal scaling normalization
df['Salary_decimal_scaled'] = df['Salary'] / (10 ** j)
df.head()
```

# Max Normalization

```python
df['Salary_max_norm'] = df['Salary'] / df['Salary'].max()

df.head()
```
