# Methods of Outlier Detection

# IQR

```python
import numpy as np
data =np.array([10,12,11,13,12,100])
q1 = np.percentile(data,25)
q3 = np.percentile(data,75)
IQR = q3-q1
lb = q1-1.5*IQR
ub = q3+1.5*IQR
outliers = data[(data<lb) |(data>ub)]
print("outliers",outliers)
```

# Scatter Plot

```python
import matplotlib.pyplot as plt 
data = [10, 12, 11, 13, 12, 100] 
plt.scatter(range(len(data)), data) 
plt.show()
```

# BOXPLOT

```python
import pandas as pd 
import matplotlib.pyplot as plt 
df = pd.DataFrame({'Marks': [45, 50, 52, 48, 49, 120]}) 
df.boxplot(column='Marks') 
plt.show() 
#Output: Box plot showing 120 as an outlier 
```

# Z-Score method

```python
import numpy as np 
data = np.array([50, 51, 49, 50, 52, 48, 51, 49, 50, 52,1000 
]) 
mean = np.mean(data) 
std = np.std(data) 
print("Mean:", mean) 
print("Standard Deviation:", std) 
z = (data - mean) / std 
print("Z :", z) 
outliers = data[(z > 3) | (z < -3)] 
print(outliers)
```

# Mean and Standard Deviation Method

```python
import numpy as np  
data = np.array([10, 12, 11, 13, 12, 100])  
mean = np.mean(data)  
std = np.std(data)  
print("Mean:",mean)  
outliers = data[(data < mean - 3*std) | (data > mean + 3*std)]  
print(outliers)
```
